```python
user = 5
numbers = []

while user > 1:
    if user % 2 == 0:
        print "The number ", user, " is even."
        user /= 2
        print "The new number is ", user, "\n\n"

    else:
        print "The number ", user, " is odd."
        user = (user * 3) + 1
        print "The new number is ", user, "\n\n"
    numbers.append (user)
```
